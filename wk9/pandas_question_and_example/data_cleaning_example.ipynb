```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('googleplaystore.csv')
df.head()
```

```python
# adapted and borrowed from here:
# https://stackoverflow.com/questions/42865724/python-parse-human-readable-filesizes-into-bytes/42865957

def parseSize(size):
    if size == 'Varies with device':
        return np.NAN
    elif type(size) is not str:
        return np.NAN
    
    size_characters = []
    for letter in size:
        if letter.lower() in ('k','m','g','t'):
            size_characters.append(' ')
        size_characters.append(letter)
    size = ''.join(size_characters)
    
    try:
        if len(size.split()) != 2:
            return np.NAN
    except:
        print(size)
        print(type(size))

    
    units = {"B": 1, "K": 10**3, "M": 10**6, "G": 10**9, "T": 10**12,
            "B": 1, "k": 10**3, "m": 10**6, "g": 10**9, "t": 10**12}

    
    number, unit = [string.strip() for string in size.split()]
    return int(float(number)*units[unit])
```

Testing it out...

```python
example_strings = ["10.43K", "11G", "343.1M",'Varies with device']

for example_string in example_strings:
    print(parseSize(example_string))

```

## Let's Create a new column
### using parseSize() function
### and, diviting by 10**6 (to convert to MB, instead of bytes)

```python
df['Size_in_MB'] = df.Size.apply(parseSize)/10**6
```
